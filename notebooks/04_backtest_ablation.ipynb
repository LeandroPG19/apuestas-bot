# Backtest ablation — compara 2 modelos champion vs challenger

Walk-forward sobre mismo período, reporta ROI/CLV/Brier/log-loss y test de Wilcoxon.

```python
sport = 'nba'
champion_version = 'v3'
challenger_version = 'v4'
seasons = '2023-24,2024-25'
```

```python
import asyncio
from apuestas.ml.backtest import run_walkforward_backtest

result_a = asyncio.run(run_walkforward_backtest(sport=sport, model_version=champion_version, seasons=seasons))
result_b = asyncio.run(run_walkforward_backtest(sport=sport, model_version=challenger_version, seasons=seasons))
```

```python
from scipy.stats import wilcoxon
import numpy as np

diffs = np.array(result_b['per_bet_pnl']) - np.array(result_a['per_bet_pnl'])
stat, pvalue = wilcoxon(diffs, alternative='greater')
print(f'Champion: ROI={result_a["roi"]:.3%}  CLV={result_a["clv"]:.3%}  Brier={result_a["brier"]:.4f}')
print(f'Challenger: ROI={result_b["roi"]:.3%}  CLV={result_b["clv"]:.3%}  Brier={result_b["brier"]:.4f}')
print(f'Wilcoxon p-value (challenger > champion): {pvalue:.4f}')
print('✅ Promover' if pvalue < 0.05 else '❌ Mantener champion')
```
